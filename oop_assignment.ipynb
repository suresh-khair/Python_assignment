```python

## Question1
#####  creating a class representing a stack data structure. Include methods for pushing and popping elements


class Stack:
    def __init__(self):
        self.stack = []

    def push(self, value):
        self.stack.append(value)
        print(f"Element {value} pushed into stack.")

    def pop(self):
        if not self.is_empty():
            popped_element = self.stack.pop()
            print(f"Element {popped_element} popped from stack.")
            return popped_element
        else:
            print("Stack is empty.")
            return None

    def is_empty(self):
        return len(self.stack) == 0

# Usage
stack = Stack()
stack.push(10)
stack.push(20)
stack.pop()
stack.pop()
stack.pop()
```

```python
## Question2

####  create a class representing a linked list data structure. 
# Include methods for displaying linked list data, inserting and deleting nodes


class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insert(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
        print(f"Inserted {data} to the linked list.")

    def delete(self, data):
        if self.head is None:
            print("Linked list is empty.")
            return
        if self.head.data == data:
            self.head = self.head.next
            print(f"Deleted {data} from the linked list.")
            return
        current = self.head
        while current.next:
            if current.next.data == data:
                current.next = current.next.next
                print(f"Deleted {data} from the linked list.")
                return
            current = current.next
        print(f"{data} not found in the linked list.")

    def display(self):
        if self.head is None:
            print("Linked list is empty.")
            return
        current = self.head
        print("Linked list: ", end="")
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")


ll = LinkedList()
print("The Linked List is:")
ll.display()

ll.insert(10)
ll.insert(20)
ll.insert(30)
print("Inserted Linked List :")
ll.display()

ll.delete(20)
print("Linked list after deleting an existing node:")
ll.display()
```

```python
## Question3

## create a class representing a shopping cart. 
# Include methods for adding and removing items, and calculating the total price.

class ShoppingCart:
    def __init__(self):
        self.cart = {}

    def add_item(self, item_name, price, quantity=1):
        if item_name in self.cart:
            self.cart[item_name]['quantity'] += quantity
        else:
            self.cart[item_name] = {'price': price, 'quantity': quantity}
        print(f"Added {quantity} of {item_name} to the cart.")

    def remove_item(self, item_name):
        if item_name in self.cart:
            del self.cart[item_name]
            print(f"Removed {item_name} from the cart.")
        else:
            print(f"{item_name} not found in the cart.")

    def calculate_total(self):
        total = sum(item['price'] * item['quantity'] for item in self.cart.values())
        return total

    def display_cart(self):
        if not self.cart:
            print("Cart is empty.")
        else:
            print("Shopping Cart:")
            for item_name, details in self.cart.items():
                total_price = details['quantity'] * details['price']
                print(f"{item_name}: {details['quantity']} quantity {details['price']} price  {total_price:.2f} total")
        
        # Optionally, you can also display the overall cart total here:
        print(f"Total Price of Cart: {self.calculate_total():.2f}")

# Usage
cart = ShoppingCart()
cart.add_item("Apple", 150, 3)
cart.add_item("Banana", 50, 5)
cart.display_cart()

# Calculate total before removing an item
print(f"Total Price: {cart.calculate_total():.2f}")

# Remove an item
cart.remove_item("Apple")

# Display the cart and calculate total again after removing the item
cart.display_cart()
```
