```python
#1. Read a text file and return the number of spaces.

def count_spaces_in_file(file_path):
    with open(file_path, 'r') as file:
        content = file.read()
        return content.count(' ')

# Example usage:
file_path = 'SampleTextFile.txt'
print(f"Number of spaces: {count_spaces_in_file(file_path)}")
```

```python
# 2. Read a text file and replace all the vowels with uppercase.

def replace_vowels_with_capital(file_path):
    vowels = 'aeiou'
    with open(file_path, 'r') as file:
        content = file.read()
    modified_content = ''.join([char.upper() if char in vowels else char for char in content])
    
    with open(file_path, 'w') as file:  # Optionally write back to the file
        file.write(modified_content)
    
    return modified_content

# Example usage:
file_path = 'SampleTextFile.txt'
print(replace_vowels_with_capital(file_path))
```

```python
# 3. Display the files in a directory.

import os

def list_files_in_directory(directory):
    return os.listdir(directory)

# Example usage:
directory_path = './'
print(list_files_in_directory(directory_path))
```

```python
# 4. Get File Creation and Modification DateTime of a file.

import os
import time

def get_file_creation_modification_time(file_path):
    creation_time = time.ctime(os.path.getctime(file_path))
    modification_time = time.ctime(os.path.getmtime(file_path))
    
    return creation_time, modification_time

# Example usage:
file_path = 'SampleTextFile.txt'
creation_time, modification_time = get_file_creation_modification_time(file_path)
print(f"Creation time: {creation_time}, Modification time: {modification_time}")
```

```python
# 5 

# A) Total number of JSON records in the file.

import json

def count_json_records(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
    return len(data)

file_path = 'Persons1.json'
print(f"Total JSON records: {count_json_records(file_path)}")
```

```python
# B) How many persons have blue eyes.

def count_blue_eyed_people(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
    return sum(1 for person in data if person.get('eyeColor') == 'blue')

file_path = 'Persons1.json'
print(f"People with blue eyes: {count_blue_eyed_people(file_path)}")
```

```python
# C) How many females and males are there.

def count_gender(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
    male_count = sum(1 for person in data if person.get('gender') == 'male')
    female_count = sum(1 for person in data if person.get('gender') == 'female')
    return male_count, female_count

file_path = 'Persons1.json'
male, female = count_gender(file_path)
print(f"Males: {male}, Females: {female}")
```

```python
# D) How many people work in companies located in France.

def count_people_in_france(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
    return sum(1 for person in data if person.get('company', {}).get('location') == 'France')

file_path = 'Persons1.json'
print(f"People working in France: {count_people_in_france(file_path)}")
```

```python
# E) Display the name of the person, registered in the following format:

from datetime import datetime

def display_registration_info(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)

    for person in data:
        registration_date = person.get('registered')
        if registration_date:
            # Parse the date string and format it
            date_obj = datetime.strptime(registration_date, "%Y-%m-%dT%H:%M:%S%z")
            formatted_date = date_obj.strftime("%a %d-%b-%Y").lower()  # Formatting as 'wed 11-feb-2015'
            # Convert the month to lowercase and print the name with formatted date
            print(f"Name: {person['name']}, Registered on: {formatted_date}")

file_path = 'Persons1.json'
display_registration_info(file_path)
```
